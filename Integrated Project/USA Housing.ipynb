```python
import pandas as pd
```

```python
hd = pd.read_csv('USA_Housing.csv')
```

```python
hd
```

```python
hd.info()
```

```python
hd.drop('Address',axis=1,inplace = True)
```

```python
hd.info()
```

### Spliting x and y

```python
y = hd['Price']
```

```python
x = hd.drop('Price',axis = 1)
```

```python
hd
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Martins_lin_model = LinearRegression()
```

```python
Martins_lin_model.fit(x_train,y_train)
```

```python
Martins_lin_model.coef_
```

```python
Martins_lin_model.intercept_
```

```python
predictions = Martins_lin_model.predict(x_test)
```

```python
result = pd.DataFrame(predictions,columns=['Prediction'])
```

```python
result
new_y_test = y_test.reset_index(drop=True)
```

```python
result['Y_test'] = new_y_test
```

```python
result['Error'] = result['Prediction'] - result['Y_test']
```

```python
result
```

```python
result_co = pd.DataFrame(Martins_lin_model.coef_, columns=['Coeff'])
```

```python
x_train.columns
```

```python
result_co['Features'] = x_train.columns
```

```python
result_co
```
